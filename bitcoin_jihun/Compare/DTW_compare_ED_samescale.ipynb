```python
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
fs = 20
t = np.arange(0, 3, 1 / fs)
f = 30
arraysA = np.sin(4*t) 
arraysB = -np.sin(4*t)
```

```python
distance_eu = euclidean(arraysA, arraysB)
distance_dtw, path = fastdtw(arraysA, arraysB, dist=euclidean)
```

```python
plt.figure(figsize=(16,5),dpi=100)
plt.plot(range(len(arraysA)),arraysA, color='red')
plt.plot(range(len(arraysB)),arraysB, color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("DTW distance same scale | dtw distance: {} vs Eucliean distance: {}".format(round(distance_dtw, 3), round(distance_eu, 3)), fontsize=15)
plt.legend(['sin(4*t)','-sint(4*t)'])
plt.show()
print("Corrcoef")
np.corrcoef(arraysA, arraysB)
```
