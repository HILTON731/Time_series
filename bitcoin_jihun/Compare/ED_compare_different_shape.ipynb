```python
from math import *
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
from scipy.spatial import distance
```

```python
data_path = "C:/Users/tony9/OneDrive/문서/DKE_assignment/Time series/time-series-analysis/bitcoin_jihun/DTW/dataset/"
file_list = os.listdir(data_path)
file_list_csv = [file for file in file_list if file.endswith(".csv")]

arraysA = pd.read_csv("C:/Users/tony9/OneDrive/문서/DKE_assignment/Time series/time-series-analysis/bitcoin_jihun/DTW/dataset/{}".format(file_list_csv[8]))
arraysB = pd.read_csv("C:/Users/tony9/OneDrive/문서/DKE_assignment/Time series/time-series-analysis/bitcoin_jihun/DTW/dataset/{}".format('HymanMinsky.csv'))

arraysA['Date'] = arraysA['Date'].map(pd.to_datetime)
arraysB.set_index(arraysB['x'])

arraysA = np.array(arraysA['Close']).reshape(-1)
arraysB = np.array(arraysB['y']).reshape(-1)
```

```python
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2)for a, b in zip(x,y)))
```

```python
#dist_scipy = distance.euclidean(arraysA, arraysB)
distance = euclidean_distance(arraysA, arraysB)
```

```python
plt.figure(figsize=(16,5),dpi=100)
plt.plot(range(len(arraysA)),arraysA, color='red')
plt.plot(range(len(arraysB)),arraysB, color='blue')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Euclidean distance Hyminsky vs Bitcoin {}| distance: {}".format(file_list_csv[8], np.round(distance, 3)), fontsize=15)
plt.legend(['Bitcoin','Hyminsky'])
plt.show()
```

```python
interp_ind = np.arange(0, len(arraysA),len(arraysA)/float(len(arraysB)))
arraysB = np.interp(np.arange(0,len(arraysA),1),interp_ind,arraysB)
```

```python
plt.figure(figsize=(16,5),dpi=100)
plt.plot(range(len(arraysA)),arraysA, color='red')
plt.plot(range(len(arraysB)),arraysB, color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euclidean distance different shape | distance: {}".format(round(distance, 3)), fontsize=15)
plt.legend(['sin(4*t1)','sint(4*t2)'])
plt.show()
print("상관계수")
np.corrcoef(arraysA, arraysB)
```
