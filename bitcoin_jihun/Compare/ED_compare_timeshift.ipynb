```python
from math import *
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
from scipy.spatial import distance
```

```python
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2)for a, b in zip(x,y)))
```

```python
fs = 20
t = np.arange(0, 3, 1 / fs)
f = 30
arraysA = np.sin(4*t) 
arraysB = np.sin(4*t+3) 
```

```python
distance = euclidean_distance(arraysA, arraysB)
```

```python
plt.figure(figsize=(16,5),dpi=100)
plt.plot(range(len(arraysA)),arraysA, color='red')
plt.plot(range(len(arraysB)),arraysB, color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euclidean distance Timeshift | distance: {}".format(round(distance, 3)), fontsize=15)
plt.legend(['sin(4*t)','sint(4*t+3)'])
plt.show()
```
